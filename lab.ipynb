```python
tokens=x.lower().split()
#type is a specific inflection of a word or token
#"maintained" is a type which is a unique form of the word, counted as one - maintained as a token is the number of all maintained
from collections import Counter
Counter(tokens)
```

```python
len(Counter(tokens))
#tells me how many types or unique word forms
len(tokens)
#tells me how many tokens (in this case, how many words)
```
